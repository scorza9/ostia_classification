```python
# image from CAT08
cat08_image_file = os.path.join(
    cat08_dataset_folder,
    DATASET_CAT08_IMAGES[0]
)

image_cat08 = Cat08ImageCT(cat08_image_file)

#print(image.name)
print("DATA SHAPE: ",image_cat08.data.shape) # the actual CT data in (i, j, k) (i ~ x, k ~ y, k ~ z)
print("\nBOUNDING BOX: ",image_cat08.bounding_box)
#print(image.origin)     # In the RAS coordinate system, this is the origin of the image
print("\nSPACING: ",image_cat08.spacing)    # Pixel spacing in the x, y and z directions (in mm)

# to see all ythe available affines, start writing image.affine_
# and check out what the autocompletion tool shows you.
# The most important one is:
print("\nAFFINE CENERLINES2RAS:\n",image_cat08.affine_centerlines2ras)
```

Images have 512x512 pixels, the number of slices depends on the number of patients.

```python
# check images asoca

fig, ax = plt.subplots(1, 1, subplot_kw={"projection": "3d"})
xl, yl, zl = image_asoca.bounding_box.get_xlim(), image_asoca.bounding_box.get_ylim(), image_asoca.bounding_box.get_zlim()
ax.set_xlim([xl[0]-10, xl[1]+10])
ax.set_ylim([yl[0]-10, yl[1]+10])
ax.set_zlim([zl[0]-10, zl[1]+10])
ax.add_artist(image_asoca.bounding_box.get_artist())
ax.scatter(image_asoca.origin[0], image_asoca.origin[1],image_asoca.origin[2], c="r")
# get image samples # lower
for zs in [-180, -140, -80]:
  points_to_sample = []
  for xs in range(-110, 110, 2):
    for ys in range(-70, 130, 2):
      points_to_sample.append([xs, ys, zs])
  points_to_sample = numpy.array(points_to_sample) # N x 3
  samples = image_asoca.sample(points_to_sample.T, interpolation="nearest")
  # plot them
  ax.scatter(points_to_sample[:,0], points_to_sample[:,1], points_to_sample[:,2], c=samples, cmap="gray")
plt.title("ASOCA")
plt.show()
```

```python
# check graphs

# asoca
asoca_graph_file = os.path.join(
    asoca_dataset_folder,
    DATASET_ASOCA_GRAPHS_RESAMPLED_05MM_DICT["Normal"][0] # "Normal", or "Diseased", from 0 to 19
)

# cat08
cat08_graph_file = os.path.join(
    cat08_dataset_folder,
    DATASET_CAT08_GRAPHS[0]
)

# load and view graphs
graph_asoca = hcatnetwork.io.load_graph(
    file_path=asoca_graph_file,
    output_type=hcatnetwork.graph.SimpleCenterlineGraph
)

graph_cat08 = hcatnetwork.io.load_graph(
    file_path=cat08_graph_file,
    output_type=hcatnetwork.graph.SimpleCenterlineGraph
)
# plot asoca graph
hcatnetwork.draw.draw_simple_centerlines_graph_2d(graph_asoca, backend="networkx")
# plot cat08 graph
hcatnetwork.draw.draw_simple_centerlines_graph_2d(graph_cat08, backend="networkx")
```
